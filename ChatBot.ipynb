```python
# I have downloaded the "new1.txt" which was a test file, so I used the file handling tools like 'file.readlines()' 
# Algorithm
# 1. iterate over all lines
# 2. check if line contains the keyword 'Is'
# 3. if true. then its a question line. So split around '?' the first part being the question and the second part being the answer
# 4. if false then its a normal statement. if the previous line was question then append this line as a new entry in the list
# 5. if previous line was also statement then append this line to the last item in the list (club statements together)
```

```python
import regex as re 
import pickle as pkl

file1 = open('new1.txt', 'r')
Lines = file1.readlines()
res = ['']
flag = True
for line in Lines:
    if flag and 'Is' not in line:
        res[-1] += line.split(' ', maxsplit=1)[1].replace('\n', ' ').strip()
    elif 'Is' in line:
        m = line.split('?')
        res.append(m[0].split(' ', maxsplit=1)[1] + '?')
        res.append(m[1].split(' ')[1].strip())
        flag = False
    else:
        res.append(line.split(' ', maxsplit=1)[1].replace('\n', ' ').strip())
        flag = True
m = []
k = 0
temp = []
for line in res:
    L = re.split('\s+|(\W)', line)
    temp.append([x for x in L if x is not None and x!=''])
    k += 1
    if k == 3:
        temp[-1] = temp[-1][0]
        m.append(tuple(temp))
        temp = []
        k = 0
print(len(m))
pkl.dump(m, open('something_test.pkl', 'wb'))
jk = pkl.load(open('something_test.pkl', 'rb'))
print(len(jk))
```

```python
# importing the deep learning libraries
from tensorflow.keras.preprocessing.sequence import pad_sequences
from keras.preprocessing.text import Tokenizer
```

```python
# tokeniser will help to provide a unique Id number to each word and punctuation marks.
# here the filters is kept as empty since we want all the punctuation and spaces to be included
tokeniser = Tokenizer(filters=[])
```

```python
# tokeniser is implemented on the vocab using the 'fit_on _text(vocab)'
tokeniser.fit_on_texts(vocab)
```

```python
# here we have created a function that vectorizes the story sequences
def vectorize_stories(data, word_index = tokeniser.word_index,
                      max_len_story = max_len_story, max_len_question = max_len_question):
    X = []
    Xq = []
    Y = []
    
    for (story, querry, answer) in data:
        x = [word_index[word.lower()]for word in story]
        xq = [word_index[word.lower()]for word in querry]
        y = [1] if tokeniser.word_index[answer]==tokeniser.word_index['yes'] else [0]
        
        X.append(x)
        Xq.append(xq)
        Y.append(y)
    return(pad_sequences(X, maxlen = max_len_story),
          pad_sequences(Xq, maxlen = max_len_question),
          np.array(Y))
   
```

```python
# here we have stories, question and answer from the vectorize_stories function for the train data
inputs_train, queries_train, answer_train = vectorize_stories(train_data)
```

```python
# importing the deep learning libraries
from keras.models import Sequential, Model
from tensorflow.keras.layers import Embedding
from keras.layers import Input, Activation, Dense, Permute, Dropout,add,dot, concatenate,LSTM
```

```python
input_sequences = Input((max_len_story,))
question = Input((max_len_question))
```

```python
# here we are adding the layers
input_encoder_m = Sequential()
input_encoder_m.add(Embedding(input_dim = vocab_len,output_dim = 64))
input_encoder_m.add(Dropout(0.3))
```

```python
input_encoder_c = Sequential()
input_encoder_c.add(Embedding(input_dim = vocab_len,output_dim = max_len_question))
input_encoder_c.add(Dropout(0.3))
```

```python
question_encoder = Sequential()
question_encoder.add(Embedding(input_dim = vocab_len, output_dim = 64, input_length = max_len_question))
question_encoder.add(Dropout(0.3))
```

```python
# encoding the seqiences
input_encoded_m = input_encoder_m(input_sequences)
input_encoded_c = input_encoder_c(input_sequences)
question_encoded = question_encoder(question)
```

```python
match = dot([input_encoded_m,question_encoded], axes = (2,2))
match = Activation('sigmoid')(match)
```

```python
response = add([match,input_encoded_c])
response = Permute((2,1))(response)
```

```python
# cocatenate
answer = concatenate([response,question_encoded])
```

```python
answer
```

```python
answer = LSTM(32)(answer)
```

```python
answer
```

```python
answer = Dropout(0.5)(answer)
answer = Dense(32)(answer)
answer = Dense(1)(answer)
```

```python
# adding the Activation function
answer = Activation('sigmoid')(answer)
```

```python
model = Model([input_sequences, question],answer)
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# for displaying the summary of the overall model
model.summary()
```

```python
history = model.fit([inputs_train,queries_train], answer_train,
                   batch_size = 32, epochs = 50,
                   validation_data = ([inputs_test, queries_test], answer_test)
                   )
```

```python
# here we plotting the the graph of Model accuracy
import matplotlib.pyplot as plt
print(history.history.keys())
plt.plot(history.history['accuracy'])

plt.plot(history.history['val_accuracy'])
plt.title("Model Accuracy")
plt.ylabel("Accuracy")
plt.xlabel("epochs")
plt.legend(["Accuracy","Val_Accuracy"])
```
